```python
import warnings
warnings.filterwarnings("ignore")
import urllib3
from bs4 import BeautifulSoup
import pandas as pd
#import geopy

http = urllib3.PoolManager()
```

```python
#shows all classifiesr with a 'predict_proba'
from sklearn.utils.testing import all_estimators


estimators = all_estimators(include_meta_estimators=True, include_other=True)
nestimators = len(estimators)
est_list = [estimator[0] for estimator in estimators]
predict_proba_list = [""] * nestimators
feature_importances__list = [""] * nestimators
predict_log_proba_list = [""] * nestimators
staged_decision_function_list = [""] * nestimators

i=0
for name, class_ in estimators:
    if hasattr(class_, 'predict_proba'):
            predict_proba_list[i] = True
    else:
        predict_proba_list[i] = False
        
    if hasattr(class_, 'feature_importances_'):
            feature_importances__list[i] = True
    else:
        feature_importances__list[i] = False
        
    if hasattr(class_, 'predict_log_proba'):
            predict_log_proba_list[i] = True
    else:
        predict_log_proba_list[i] = False
        
    if hasattr(class_, 'staged_decision_function'):
            staged_decision_function_list[i] = True
    else:
        staged_decision_function_list[i] = False
    i += 1
            
        
feat = [('Predict_proba', predict_proba_list),
         ('Feature Importance', feature_importances__list),
        ('predict_log_proba', predict_log_proba_list),
        ('staged_decision_function', staged_decision_function_list)
       ]
    
df = pd.DataFrame.from_items(feat)
df.index=est_list
        
```

```python
df.shape
```

```python

classifiers_=[('classifier',*all_estimators(type_filter= 'classifier')[i]) for i in range(0,len(all_estimators(type_filter= 'classifier')))]
regressors_=[('regressor',*all_estimators(type_filter= 'regressor')[i]) for i in range(0,len(all_estimators(type_filter= 'regressor')))]
clusters_=[('cluster',*all_estimators(type_filter= 'cluster')[i]) for i in range(0,len(all_estimators(type_filter= 'cluster')))]
transformers_=[('transformer',*all_estimators(type_filter= 'transformer')[i]) for i in range(0,len(all_estimators(type_filter= 'transformer')))]
unknowns_=[('unknown',*all_estimators(type_filter= None)[i]) for i in range(0,len(all_estimators(type_filter= None)))]
```

```python
estimators__ = all_estimators(include_meta_estimators=True)
estimators___ = all_estimators(include_meta_estimators=False)
meta_estimators = list(set(estimators__)-set(estimators___))
meta_estimators_ = [('meta estimators', *meta_estimators[i]) for i in range(0, len(meta_estimators))]
meta_estimators_
```

```python
others__ = all_estimators(include_other=True)
others___ = all_estimators(include_other=False)
others = list(set(others__)-set(others___))
others_ = [('others', *others[i]) for i in range(0, len(others))]
others_
```

```python
xx = clusters_+classifiers_+regressors_+transformers_+others_+meta_estimators_
```

```python
type_list = [estimator[0] for estimator in xx]
len(type_list)
```

```python
name_list = [estimator[1] for estimator in xx]
len(name_list)
```

```python
df_ = pd.DataFrame()
df_['name'] = name_list
df_['type'] = type_list
```

```python
df_.set_index('name', inplace=True)
```

```python
df = pd.merge(df, df_ , left_index=True, right_index=True)
df.reset_index(inplace=True)
```
