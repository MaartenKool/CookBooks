```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

theta = np.linspace(0, 1*np.pi, 1000)
thetab = np.linspace(0, -1*np.pi, 1000)

radius1 = 0.3
radius2 = 0.275
radius3 = 0.25
radius4 = 0.225

a1a = radius1*np.cos(theta)
b1a = radius1*np.sin(theta)

a2a = radius2*np.cos(theta)
b2a = radius2*np.sin(theta)

a3a = radius3*np.cos(theta)
b3a = radius3*np.sin(theta)

a4a = radius4*np.cos(theta)
b4a = radius4*np.sin(theta)

a1b = radius1*np.cos(thetab)
b1b = radius1*np.sin(thetab)

a2b = radius2*np.cos(thetab)
b2b = radius2*np.sin(thetab)

a3b = radius3*np.cos(thetab)
b3b = radius3*np.sin(thetab)

a4b = radius4*np.cos(thetab)
b4b = radius4*np.sin(thetab)

figure, axes = plt.subplots(1)

axes.plot(a1a, b1a)
axes.plot(a2a, b2a)
axes.plot(a3a, b3a)
axes.plot(a4a, b4a)
# axes.plot(a1b-.025, b1b)
# axes.plot(a2b-.025, b2b)
# axes.plot(a3b-.025, b3b)
axes.plot(a4b-.025, b4b)




axes.set_aspect(1)

plt.title('Circle using Parametric Equation')
plt.show()
```

```python
# """
#  Spiral.py is a program which make a polar plot of r = theta^2
#    i.e. the Galilean spiral

#  Paul Eugenio
#  Florida State University
#  PHZ4151C
#  29 Jan 2019
# """

from __future__ import division, print_function
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
# use latex 
plt.rc('text', usetex=False)

#  define parameters
thetaMin, thetaMax = 26*np.pi, 36*np.pi
steps = 223


# use numpy vectorization to generate plotting values
theta = np.linspace(thetaMin, thetaMax, steps)
r = theta*theta
x = r*np.cos(theta)
y = r*np.sin(-theta)
# 90*np.pi*r

fig, ax = plt.subplots(figsize=(8,8))
# ax.set(xlim=xlim, ylim=ylim, autoscale_on=False)


ax.set_aspect('equal')
# plot curve and decorate with latex equation
plt.plot(x, y)
# plt.plot(x, y)
plt.xlabel(r"$r(\theta)$")
plt.ylabel(r"$\theta^2$")
plt.title("Galilean spiral")
# plt.savefig("spiral.png")
plt.show()
```
