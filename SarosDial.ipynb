```python

import numpy as np
import matplotlib.pyplot as plt
import math

def fermat_spiral(dot):
    data=[]    
    d=dot*0.1
    for i in range(dot): 
        t = i / d * np.pi
        x = (1 +  t) * math.cos(t)
        y = (1 +  t) * math.sin(t)
        data.append([x,y])
    narr = np.array(data)
    f_s = np.concatenate((narr,-narr))
    return f_s

f_spiral = fermat_spiral(2000)
plt.scatter(f_spiral[len(f_spiral)//2:,0],f_spiral[len(f_spiral)//2:,1])
# plt.scatter(f_spiral[:len(f_spiral)//2,0],f_spiral[:len(f_spiral)//2,1])
plt.show()
```

```python
# """
#  Spiral.py is a program which make a polar plot of r = theta^2
#    i.e. the Galilean spiral

#  Paul Eugenio
#  Florida State University
#  PHZ4151C
#  29 Jan 2019
# """

from __future__ import division, print_function
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
# use latex 
plt.rc('text', usetex=False)

# define parameters
thetaMin, thetaMax = 0, 20*np.pi
steps = 1000


# use numpy vectorization to generate plotting values
theta = np.linspace(thetaMin, thetaMax, steps)
r = theta*theta
x = r*np.cos(theta)
y = r*np.sin(-theta)
# 90*np.pi*r

fig, ax = plt.subplots(figsize=(8,8))
# ax.set(xlim=xlim, ylim=ylim, autoscale_on=False)


ax.set_aspect('equal')
# plot curve and decorate with latex equation
plt.plot(x[600:], y[600:])
# plt.plot(x, y)
plt.xlabel(r"$r(\theta)$")
plt.ylabel(r"$\theta^2$")
plt.title("Galilean spiral")
# plt.savefig("spiral.png")
plt.show()
```

```python
#https://stackoverflow.com/questions/53232147/thinkpython-ex-4-5-write-a-program-that-draws-an-archimedes-spiral-or-one-o
import turtle
def spiral(turtle, rotations=6, a=0.0, b=5):
    theta = 0.0

    while theta < rotations * 2 * pi:
        radius = a + b * theta
        x, y = radius * cos(theta), radius * sin(theta)
        turtle.goto(x, y)
        theta += 0.1
```
