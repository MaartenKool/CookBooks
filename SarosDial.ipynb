```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

theta = np.linspace(0, 1*np.pi, 1000)
thetab = np.linspace(0, -1*np.pi, 1000)

radius1 = 0.3
radius2 = 0.275
radius3 = 0.25
radius4 = 0.225

a1a = radius1*np.cos(theta)
b1a = radius1*np.sin(theta)

a2a = radius2*np.cos(theta)
b2a = radius2*np.sin(theta)

a3a = radius3*np.cos(theta)
b3a = radius3*np.sin(theta)

a4a = radius4*np.cos(theta)
b4a = radius4*np.sin(theta)

a1b = radius1*np.cos(thetab)
b1b = radius1*np.sin(thetab)

a2b = radius2*np.cos(thetab)
b2b = radius2*np.sin(thetab)

a3b = radius3*np.cos(thetab)
b3b = radius3*np.sin(thetab)

a4b = radius4*np.cos(thetab)
b4b = radius4*np.sin(thetab)

figure, axes = plt.subplots(1)

axes.plot(a1a, b1a)
axes.plot(a2a, b2a)
axes.plot(a3a, b3a)
axes.plot(a4a, b4a)
# axes.plot(a1b-.025, b1b)
# axes.plot(a2b-.025, b2b)
# axes.plot(a3b-.025, b3b)
axes.plot(a4b-.025, b4b)




axes.set_aspect(1)

plt.title('Circle using Parametric Equation')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#As a path collection

from matplotlib.collections import LineCollection

fig, ax = plt.subplots()
r = np.arange(0, .075, 0.00001)
nverts = len(r)
theta = np.array(range(nverts)) * (2*np.pi)/(nverts-1)
theta = 90*np.pi*r
xoffset, yoffset = .0, .0
yy = 2*r * np.sin(theta) + yoffset
xx = r * np.cos(theta) + xoffset
spiral = zip(xx,yy)
collection = LineCollection([list(spiral)], colors='k')
ax.add_collection(collection)
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)

# In polar coordinates

fig, ax = plt.subplots(subplot_kw=dict(polar=True), figsize=(.5,.5))
r = np.arange(0, 3.4, 0.01)
theta = 2*np.pi*r
# ax.set_axis_off()
# ax.grid(False)
# ax.axes.get_xaxis().set_visible(False)
# ax.axes.get_yaxis().set_visible(False)
ax.plot(theta, r, linewidth=1, color='k');
```

```python

import numpy as np
import matplotlib.pyplot as plt
import math

def fermat_spiral(dot):
    data=[]    
    d=dot*0.1
    for i in range(dot): 
        t = i / d * np.pi
        x = (1 +  t) * math.cos(t)
        y = (1 +  t) * math.sin(t)
        data.append([x,y])
    narr = np.array(data)
    f_s = np.concatenate((narr,-narr))
    return f_s

f_spiral = fermat_spiral(2000)
plt.scatter(f_spiral[len(f_spiral)//2:,0],f_spiral[len(f_spiral)//2:,1])
# plt.scatter(f_spiral[:len(f_spiral)//2,0],f_spiral[:len(f_spiral)//2,1])
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib.cm as cm 

n = 256
angle = np.linspace(0,12*2*np.pi, n)
radius = np.linspace(.5,1.,n)

x = radius * np.cos(angle)
y = radius * np.sin(angle)

plt.scatter(x,y,c = angle, cmap = cm.hsv)
plt.show()
```

```python
# """
#  Spiral.py is a program which make a polar plot of r = theta^2
#    i.e. the Galilean spiral

#  Paul Eugenio
#  Florida State University
#  PHZ4151C
#  29 Jan 2019
# """

from __future__ import division, print_function
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
# use latex 
# plt.rc('text', usetex=True)

# define parameters
thetaMin, thetaMax = 0, 10*np.pi
steps = 100


# use numpy vectorization to generate plotting values
theta = np.linspace(thetaMin, thetaMax, steps)
r = theta*theta
x = r*np.cos(theta)
y = r*np.sin(-theta)

# plot curve and decorate with latex equation
plt.plot(x[:90], y[:90])
# plt.plot(x, y)
plt.xlabel(r"$r(\theta)$")
plt.ylabel(r"$\theta^2$")
plt.title("Galilean spiral")
# plt.savefig("spiral.png")
plt.show()
```
