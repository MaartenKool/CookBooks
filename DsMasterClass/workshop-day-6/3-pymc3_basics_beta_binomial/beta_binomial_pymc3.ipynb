## Sampling from a pymc3 distribution

Pymc3 is conventionally imported as `pm`. We can use it for sampling from distributions pretty much in the same way as `.rvs()` in any scipy distribution object. 

Let's define suitable parameters for a beta distribution:

```python
alpha = 5
beta = 3
```

We set up the corresponding pymc3 distribution object. Typically we will work in a model context. 

```python
with pm.Model() as model:
    p = pm.Beta('p', alpha=alpha, beta=beta)
```

We can refer to the same model context and add new ingredients. Now we want to sample from the defined distribution. The sampling results will be saved in a trace, which for now we literally call like that.

```python
with model:
    trace = pm.sample(draws=5000)
```

The trace object contains the values of all the draws, by default across two chains.
We usually run several chains from different starting points in parallel. In the end we combine them. Comparing them can give diagnostic measure about the sampling procedure. The sampling adds 500 tuning steps by default. Sometimes that is more than enough, but more often it might not be enough and pymc3 will suggest to increase the number of tuning steps.

```python
trace
```

Below are the variables contained in the trace. There are more than we actually set up. For numerical convenience all distributions with support on finite intervals are transformed to distributions with support on the whole real line before sampling - that is usually only for internal use and we are not concerned about the transformed values.

```python
trace.varnames
```

We can access the results from the different iteration steps or for a specific variable.

```python
trace[0]
```

```python
trace['p'][:10]
```

We can even transform the trace to dataframe format:

```python
pm.trace_to_dataframe(trace).head()
```

The number of samples is the number of draws times the number of chains.

```python
trace['p'].shape
```

Let's compare the result with the theoretical shape of the distribution.

```python
dist = stats.beta(alpha, beta)
xvals = np.linspace(0, 1, 100)
yvals = dist.pdf(xvals)

plt.hist(trace['p'], density=True, bins=20)
plt.plot(xvals, yvals, lw=4)
plt.xlabel('p')
plt.show()
```

## Create some Bernoulli trials

We set up a prior distribution. Now we would also like to have some data, define the likelihood and obtain the posterior distribution.

```python
# define the Bernoulli distribution
prob = 0.4
bernoulli = stats.bernoulli(prob)
# number of trials
n = 25
# create random samples
sample = bernoulli.rvs(size=n, random_state=2)
# number of ones contained in sample
k = sum(sample)
print(sample)
print(np.unique(sample, return_counts=True))
```

```python
np.mean(sample)
```

### With Bernoulli likelihood

We perform the same setup, but now include the likelihood function corresponding to our observed Bernoulli trials. For the likelihood, we just set up another distribution object. The crucial difference is however that it depends on the `observed` data and on the prior distribution.

```python
alpha = 5
beta = 3
```

```python
with pm.Model() as model:
    p = pm.Beta('p', alpha=alpha, beta=beta)
    likelihood = pm.Bernoulli('likelihood', p=p, observed=sample)
```

```python
with model:
    trace = pm.sample(draws=5000, tune=1000)
```

We can ask for the definition of the model that we set up, the distributions and their
parameter dependencies.

```python
model.model
```

The dependencies can be depicted as a graphical model highlighting the dependencies and how many components each ingredient has (e.g. the number of observations).

```python
pm.model_to_graphviz(model)
```

The trace reports the posterior distribution of $p$.

```python
trace.varnames
```

Compare the trace to the algebraically derived result:

```python
dist = stats.beta(alpha, beta)
xvals = np.linspace(0, 1, 100)
yvals = dist.pdf(xvals)

dist_p = stats.beta(alpha+k, beta+n-k)
yvals_p = dist_p.pdf(xvals)

plt.hist(trace['p'], density=True, bins=20)
plt.plot(xvals, yvals, lw=4, label='prior')
plt.plot(xvals, yvals_p, lw=4, label='posterior')
plt.xlabel('p')
plt.legend()
plt.show()
```

We can obtain the mean and standard deviation of the posterior distribution.

```python
np.mean(trace['p']), np.std(trace['p'])
```

After sampling, we always have full access to the trace object and we could directly work with it. However, the benefit of using pymc3 is that it bundles a lot of functionality for analyzing the trace.


The trace summary contains the mean and standard deviation of all the (untransformed) variables of the trace. Additionally there are estimates about  the effective sample size, 
the accuracy of the Monte Carlo estimate,
the Gelman-Rubin statistic called `Rhat` or $\hat{R}$,
and the highest posterior density interval or region (HPD).

The HPD is interesting for us - it tell us where most of the posterior probability density is contained. It is different from a confidence interval since we really have a probability distribution for the respective parameter, and for non-symmetric distributions it won't be symmetric with respect to the mean. Even though HPDs and confidence intervals have really different interpretations, due to central limit theorems related to the likelihood function for large sample sizes the boundaries of HPDs and confidence intervals will agree fairly well (if the CLT conditions are met).

We will look at the definitions of the other quantities later. Basically they inform us about the quality of the sampling procedure ultimately measuring how representative the obtained samples are of the posterior distribution. In particular a value of $\hat{R}$ close to one is supposed to be good. Usually, pymc3 will analyze these measures automatically and report if it has concerns.

```python
pm.summary(trace)
```

The effective sample size measures how many independent samples would be needed to obtain an estimate of the posterior as good as the one from the MCMC. Due to their autocorrelation, the samples do not represent the posterior as well as suggested by the number of samples:

$$
n_{\rm eff} = \frac{n}{1+2\sum_{k=1}^\infty {\rm acf}(k)}\ ,
$$

The accuracy of the Monte Carlo estimate is approximately

$$
{\rm mc}_{\rm error} \approx \frac{sd}{\sqrt{n_{\rm eff}}} \ ,
$$

The Gelman-Rubin statistic called `Rhat` or $\hat{R}$ is calculated from the
within-chain-variance $W$ and between-chain-variance $B$.

From the sampling procedure we obtain $m$ chains for $n$ draws obtaining values $\psi_{ij}$
with $i\in\{1,\ldots,n\}$ and $j\in\{1,\ldots,m\}$.
Then

$$
\begin{eqnarray*}
B &=& \frac{n}{m-1}\sum_{j=1}^m(\bar{\psi}_{\cdot j}-\bar{\psi}_{\cdot \cdot})^2\\
W &=& \frac{1}{m(n-1)}\sum_{j=1}^m\sum_{i=1}^n(\psi_{ij}-\bar{\psi_{\cdot j}})^2 
\end{eqnarray*}
$$
 
where each chain mean and the average of the chain means are given by

$$
\begin{eqnarray*}
\bar{\psi}_{\cdot j}&=&\frac{1}{n}\sum_{i=1}^n\psi_{ij}\\
\bar{\psi}_{\cdot\cdot}&=&\frac{1}{m}\sum_{j=1}^m\bar{\psi}_{\cdot j}
\end{eqnarray*}
$$
 
 
The variance of the samples given the data can then be estimated as 

$$
\hat{\rm var}^+(\psi|{\rm data}) = \frac{n-1}{n} W+\frac{1}{n}B 
$$

The more samples and the more chains we have, the more this expression will be dominated by $W$. How much is measured by $\hat{R}$:

$$
\hat{R} = \sqrt{\frac{\hat{\rm var}^+(\psi|{\rm data})}{W}}
$$

### Trace plot

Pymc3 contains useful plotting functionality, for example for depicting the posterior probability distribution together with the HPD and a desired statistic (e.g. mean or mode).

```python
pm.plot_posterior(trace)
plt.show()
```

We can represent the same with a traceplot giving a kernel density estimate resolved by the different chains on the left hand side and the values obtained during different iteration steps of the chains.

```python
pm.traceplot(trace)
plt.show()
```

We usually obtain several chains through sampling where each chain starts from a different value. In the final trace we combine all these samples, but looking at the individual chains serves for recognizing if the different chains represent the same distribution. If the samples at different iterations did not "mix", i.e. overlap, the sampling procedure would not have worked well. When the sampling starts, it will usually take some time before achieving good mixing. This is expected behavior, and therefore the first few chain steps are usually discarded. In the `pm.sample()` function, by default the first 500 steps will be "burned". Sometimes that is more than necessary, but more often it might not be enough and pymc3 will suggest to increase the number of tuning steps.

### With binomial likelihood

We can define exactly the same model using a binomial likelihood instead of the Bernoulli likelihood. Then the observed values are the number of ones. 

```python
with pm.Model() as model:
    p = pm.Beta('p', alpha, beta)
    likelihood = pm.Binomial('likelihood', n=len(
        sample), p=p, observed=sample.sum())
    trace = pm.sample(draws=5000)
```

### Trace plots

```python
pm.traceplot(trace)
plt.show()
```

When plotting the posterior distribution, we can also add a reference line for example to mark the true distribution parameter that we know about. Pymc3 will then tell us about the probability mass on either side of that value. Given the data that we have we cannot completely exclude the true value.

Additionally we can define a "region of practical equivalence" (ROPE). Since we cannot expect the observed data to reproduce exactly the same parameter value, we can define through the ROPE what kind of level of fluctuations we would expect. Choosing the appropriate size of the ROPE requires domain knowledge. The ROPE may or may not overlap with the HPD, and if it does, the HPD will either be contained entirely within the ROPE or only partially. In this way, we can perform a Bayesian analog of a null hypothesis test and would come to the following decisions:

- no significant overlap between ROPE and HPD: not enough support for the null value
- HPD contained in ROPE: support for null value (even if HPD does not include null value)
- overlap between HPD and ROPE or ROPE contained in HPD: not yet able to confirm

```python
rope = np.around([prob-0.05, prob+0.05], 2)
pm.plot_posterior(trace, ref_val=prob, rope=rope,
                  point_estimate='mode', kde_plot=False)
plt.show()
```

Alternatively we can plot with a kernel density estimate.

```python
pm.plot_posterior(trace, ref_val=0.6, rope=rope,
                  point_estimate='mode', kde_plot=True)
plt.show()
```

### MCMC in a nutshell
To understand some of the further diagnostics we have to understand how the sampling procedure works. 

- First, a value is sampled from the prior distribution
- This value is inserted into the likelihood function 
- We calculate the product of the prior and likelihood function
- We normalize with the marginal likelihood and obtain the posterior probability density for the sample value

The first three steps are straight-forward. It is the last one which creates trouble. In the current very simple example we would be able to obtain the marginal, but not so if many parameters are involved. The Markov chain procedure is about avoiding performing the integral. 
Instead we use a method which obtains the proportions of the posterior purely by sampling, and that in such a way that we have to sample as small amounts of points as possible before getting a decent representation of the posterior.


- We carry out the first three steps from above
- We carry out the same three steps sampling a second value

Then we have two values $p_1$ and $p_2$ for which we calculate

$$
\begin{eqnarray*}
p(p_1|data)&\propto& p(data|p_1)p(p_1)\\
p(p_2|data)&\propto& p(data|p_2)p(p_2)
\end{eqnarray*}
$$

and we could do that for a lot more values. If we were adding all the sampled values to the chain, we would obtain the prior distribution in the end. However not all sampled values are going to be added to the trace, some of them are rather rejected. In the Metropolis-Hastings algorithm, the first invented for that purpose, the last value in the chain is compared to the  newly sampled one. For the two of them, the ratio of the posteriors is considered:

$$
\frac{p(p_2|data)}{p(p_1|data)}
\propto 
\frac{p(data|p_2)p(p_2)}
{p(data|p_1)p(p_1)}
$$

- If this ratio is larger than one, $p_2$ has higher posterior probability density than $p_1$ and is added to the chain at the next iteration step.
- If the ratio is smaller than one, we sample a random number from the continuous uniform distribution on the interval $[0,1]$ to obtain a number $\theta$. If the ratio is larger than $\theta$, $p_2$ will be added to the chain, if not, $p_1$ will be added to the chain. 
- A new proposal from the vicinity of the last proposal (generated e.g. from a normal distribution centered on the last value) is compared against the last value in the chain.


### Autocorrelations

Continuing in this way, one can show that after an infinite number of iteration steps, the posterior distribution is correctly represented. Less steps can already be useful, but since consecutive values are related to each other, there will be autocorrelations within the trace. 

```python
pm.autocorrplot(trace)
plt.show()
```

Because of the autocorrelations, we effectively explored the parameter space less well than suggested by the number of samples. That is why $n_{\rm eff}$ can be considerably lower than the number of samples, the more so the more difficulties the sampling procedure has in accepting new values and move away from previous values. Many more algorithms have been designed to make the sampling procedure more efficient. Usually pymc3 will select the best suited one automatically. The default one is called the No-Uturn-Sampler (NUTS).

You will sometimes see that traces are thinned by e.g. retaining only every third sample. This does reduce autocorrelation, but it also reduces the number of effective samples and therefore does not really address the problem.

The sampling procedure is easily parallelizable. Chains can be run separately and combined at a later stage. 

### Forest plot

A quick overview of the results can be produced with a forest plot. It depicts a chosen statistic (like the mean), the interquartile range and the HPD of the trace for each parameter and each chain. This can serve as a quick overview especially if many parameters are involved. 
Additionally it plots the Gelman-Rubin statistic for the chains of each variable. Values close to one are good.

```python
pm.forestplot(trace)
plt.show()
```

### Posterior predictive samples

Once we have obtained the posterior distribution represented through the trace, we can sample from the posterior. This will produce the predicted values based on each of the parameter combinations drawn from the posterior distribution (represented through the trace).

## Step algorithms

Different MCMC algorithms might work differently well on different problems. Let's compare some of the main characteristics:

```python
trace_test = {}

with model:
    for step in [pm.Metropolis(), pm.Slice(), pm.NUTS()]:
        trace_test[step.name] = pm.sample(draws=5000, step=step)
```

```python
pm.traceplot(trace_test['metropolis'][:100])
plt.show()
```

```python
fig, ax = plt.subplots(nrows=3, sharex=True, figsize=(12, 6))
for i, key in enumerate(trace_test.keys()):
    ax[i].plot(trace_test[key]['p'][:100], label=key)
    ax[i].legend(loc=[1.01, 0])
```

## Saving and loading traces

We can choose various types of backends where to write traces to, for example simply to text files. All we have to do is indicate a directory where the results are going to be saved, and each chain will be written to a separate file in that directory. 

There are various backend choices, for example an sqlite database is a convenient alternative.

```python
with model:
    db = pm.backends.Text('test')
    trace_ext = pm.sample(100, trace=db)
```

```python
trace_ext
```

We can the load the trace again from the external file and provide the model context.

```python
trace_load = pm.backends.text.load('test', model=model)

pm.traceplot(trace_load)
plt.show()
```

If you run the above multiple times, results will be added to the same chains. To create new chains, simply provide new chain indices with `chain_idx`.

```python
with model:
    db = pm.backends.Text('test')
    trace_ext = pm.sample(100, trace=db, chain_idx=2)
```

```python
trace_load = pm.backends.text.load('test', model=model)
trace_load
```

```python
pm.traceplot(trace_load)
plt.show()
```
