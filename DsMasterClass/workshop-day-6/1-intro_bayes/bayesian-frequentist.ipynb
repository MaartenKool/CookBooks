## Frequentist statistics

In frequentist statistics, we assume the existence of some population distribution with fixed shape parameters. We get information about the distribution of the population through samples from the population. Our goal is to determine how likely it is to observe the sample given that the population parameters have certain values:

$$
P({\rm data}|\theta)
$$

where $\theta$ is the vector of all the parameters needed to describe the distribution of the population.

Based on the observed samples we can e.g. 

- test the hypothesis that the population parameters have certain values
- obtain confidence intervals for the estimate of population statistics
- infer the population parameters through maximum likelihood fit, that is determine for which population parameters it would be most likely to observe the sample


Our estimates will only become completely accurate once we have seen the whole population.
Based on the observed samples we obtain point estimates of the population parameters. We only ever talk about probabilities of samples - we never talk about the probability of a certain population parameter.

## Bayesian statistics

In Bayesian statistics, we follow somehow the opposite direction. The data is fixed, and we determine the probability of having a certain (combination of) distribution parameter given the data:

$$
P(\theta|{\rm data})
$$

Based on the samples we can 

- assign probabilities to all the distribution parameters of the data
- determine the most probable ranges of the distribution parameters
- dynamically adjust the distribution parameters to new/different data

##Â Bayes' rule

Though the two approaches are different, both make use of joint and conditional probabilities and Bayes' rule.

The joint probability $P(A,B)$ of two events $A$ and $B$ is the probability of observing both of these events together. It can be used to define the conditional probability:

$$ P(A,B) = P(A|B) \; P(B) $$

We could equally have written 

$$ P(A,B) = P(B|A) \; P(A) $$

Equating the two expressions gives Bayes' rule

$$  P(A|B) \; P(B) = P(B|A) \; P(A) $$

Assuming that both events are possible, we can solve for either of the conditional probabilities:

$$ P(A|B) = \frac{P(B|A)\;P(A)}{P(B)} $$

## Bayes' rule for models and data

In the context of Bayesian modeling, the events $A$ and $B$ that we are concerned about are a model (consisting of parameters describing how our data should be distributed) and the observed sample of data, so that we can write:

$$P\left({\rm model}|{\rm data}\right) = \frac{P\left({\rm data}|{\rm model}\right)}{P({\rm data})}\; P\left({\rm model}\right)$$

The different parts in this equation have specific names.

$$
\begin{eqnarray*}
{\rm prior} &=& P\left({\rm model}\right)\\\\
{\rm likelihood} &=& P\left({\rm data}|{\rm model}\right)\\\\
P({\rm data}) &=& \sum_{\rm model} P({\rm data}|{\rm model}) P({\rm model})\\\\
{\rm posterior} &=& P\left({\rm model}|{\rm data}\right) 
\end{eqnarray*}
$$

- The prior distribution gives the probability of the model irrespective of the data. Prior probabilities are also referred to as prior believes indicating that they can be subjective. Usually we would assume that a prior believe will be adjusted by whatever data is observed. 

- The likelihood gives the joint probability of the data given the model. This is the object of major interest in frequentist statistics. 

- The marginal probability of the data gives the probability of observing the data across all considered models (choice of parameters and parameter values). It serves as a normalization factor. It may involve summation or integration across unbounded ranges of potentially a high number of parameters and is therefore posing challenges for numerical computation.

- The posterior gives the probability of the model given the data. It can be interpreted as taking new evidence into account which may or may not lead to model probabilities very different from the prior believes. 

Ideally the prior would take reasonable assumptions or existing knowledge into account. It allows to express opinions (founded or not) and how much data will be needed to overcome these prior believes will depend on how opionated we are. 

## Maximum likelihood estimates

The joint probability of observing a sample of independent datapoints $x_i$ is given by the product of the individual probabilities. 
Interpreting this joint probability as a function of the model parameters for a fixed dataset gives the likelihood function

$$L(\theta|x) = \prod_{i=1}^n f(x_i|\theta)=
f(x_1|\theta) \cdot f(x_2|\theta) \cdot f(x_2|\theta)...f(x_n|\theta)\ .$$

$x_1, x_2, \ldots, x_n$ are the observed values, $\theta$ stands for the model parameters (probability of having cancer, regression coefficients, etc.).

We find optimal parameter estimates by maximizing the likelihood. Often it can be easier to maximize the logarithm of the likelihood which leads to the same result.

## Example: Bernoulli trials

Consider $n$ independent Bernoulli trials, for example throwing a coin or measuring if a person has cancer or not. The Bernoulli distribution with probability $P$ (e.g. for obtaining head) is

$$f(x_i|P) = P^{x_i}(1-P)^{1-x_i}$$ 

where 

$$x_i=\begin{cases} 1 {\rm\ for\ head} \\ 0 {\rm\ for\ tail} \\ \end{cases}$$

For $n$ trials, this gives the likelihood

$$L(P|x) = \prod_{i=1}^n P^{x_i}(1-P)^{1-x_i} $$

To maximize the likelihood, we take the derivative and set it equal to 0. For the product above, that gets quickly complicated. 
So instead let's first take the logarithm:

$$ 
\begin{eqnarray}
\log L (P|x) 
&=& \log\left(\prod_{i=1}^n P^{x_i}(1-P)^{1-x_i}\right)\\
&=&\sum_{i=1}^n \log \left( P^{x_i}(1-P)^{1-x_i} \right)\\
&=&\sum_{i=1}^n \left(x_i \log(P) + (1-x_i) \log(1-P)\right)
\end{eqnarray}
$$

Now we can easily differentiate $\log L$ with respect to $P$, set the derivative equal to zero and solve for $P$. The solution we can then reinsert into $\log L$ or directly $L$. 

The result is 

$$
P = \frac{\sum_{i=1}^n x_i}{n}
$$

which in this case is not really surprising.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
sns.set(font_scale=1.5)
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
from scipy import stats
```

```python
# define the Bernoulli distribution
bernoulli = stats.bernoulli(0.4)
# number of trials
n = 25
# create random samples
sample = bernoulli.rvs(size=n, random_state=2)
# number of ones contained in sample
k = sample.sum()
print(sample)
print(np.unique(sample, return_counts=True))
```

```python
pvalues = np.linspace(0.01, 0.99, 100)
yvalues = [(sample*np.log(p_i)+(1-sample)*np.log(1-p_i)).sum()
           for p_i in pvalues]
likelihood_max = k/n
yvalues_max = k*np.log(likelihood_max)+(n-k)*np.log(1-likelihood_max)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(pvalues, yvalues)
ax.axhline(yvalues_max, 0, 1)
ax.scatter(likelihood_max, yvalues_max, color='r',
           marker='x', s=100, label='max')
ax.set_title(
    'The log-likelihood function at its maximum for n={}, k={}\n'.format(n, k))
ax.set_xlabel('p')
ax.set_ylabel('log-likelihood')
ax.legend()
plt.show()
```

### Example: Gaussian variable

We regress an outcome variable $y$ on a matrix of predictor variables $X$ and assume normally distributed residuals:

$${\rm pdf}\left(\epsilon_i=\left(y_i-(X\beta)_i\right)|\sigma\right) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{\left(y_i-(X\beta)_i\right)^2}{2\sigma^2}}$$

The likelihood function will be the product of the pdfs of all $\epsilon$:

$$
L= \frac{1}{(2\pi\sigma^2)^{\frac{n}{2}}}e^{-\frac{\left(y-X\beta\right)^T\left(y-x\beta\right)}{2\sigma^2}}
$$

The log-likelihood is 

$$
\log L = -\frac{n}{2}\log(2\pi\sigma^2)-\frac{\left(y-X\beta\right)^T\left(y-X\beta\right)}{2\sigma^2}
$$

Maximizing the log-likelihood with respect to $\beta$ leads to the regression coefficients

$$
\beta = (X^T X)^{-1}X^T y
$$

We generate points deviating by normally distributed residuals from a regression line. Then we determine the likelihood function.

```python
# number of trials
n = 25
# standard deviation
sigma = 10
# define normal distribution
normal = stats.norm(0, sigma)
# regression line for y on x
x = np.linspace(0, 10, n)
y = 2*x + normal.rvs(n, random_state=10)

# plot the related variables
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, 2*x, lw=4, color='r')
ax.scatter(x, y)
plt.show()
```

Fitting a regression line gives an estimate of the true regression coefficient.

```python
from sklearn.linear_model import LinearRegression
```

```python
# fit the linear regression model
model = LinearRegression(fit_intercept=False)
model.fit(x.reshape(-1, 1), y)
model.coef_
```

We can determine the residuals estimated based on the fitted regression line.

```python
def residuals_prod(beta, y, x, sigma):
    eps = -(y - x*beta).T.dot(y - x*beta)/(2*sigma**2)
    return (np.prod(eps))
```

```python
xvals = np.linspace(0, 4, 100)
yvals = [residuals_prod(beta, y, x, sigma) for beta in xvals]

fig, ax = plt.subplots(figsize=(8, 4))
ax.axhline(max(yvals))
ax.scatter(model.coef_, max(yvals), color='r', marker='x', s=100)
ax.plot(xvals, yvals)
ax.set_title(
    'The log-likelihood function at its maximum beta={}\n'.format(np.round(model.coef_[0], 3)))
ax.set_xlabel('beta')
ax.set_ylabel('log-likelihood')
plt.show()
```

## Example

Let's consider again [Judea Pearl's](#Pearl) example.

Consider a medical trial on two groups, 40 male and 40 female patients.
Does the drug lead to increased recovery rate? Should the drug be recommended?

|-|Recovered|Not recovered|Recovery rate|
|----|----|----|----|
|Males|
|Given drug|18|12|0.6|
|Not given drug|7|3|0.7|
|Females|
|Given drug|2|8|0.2|
|Not given drug|9|21|0.3|
|Combined|
|Given drug|20|20|0.5|
|Not given drug|16|24|0.4|


We have

$$
P(R) = \frac{36}{80} = \frac{9}{20}
$$

$$
P(D) = \frac{1}{2}
$$

$$
P(M) = \frac{1}{2}
$$

For example, we can calculate 

$$
P(R,D) = \frac{20}{80} = \frac{1}{4} 
$$

This result helps us to calculate

$$
P(D|R) = \frac{P(D,R)}{P(R)} = \frac{1}{4}\cdot\frac{20}{9} = \frac{5}{9}
$$

and 

$$
P(R|D) = \frac{P(D,R)}{P(D)} = \frac{1}{4}\cdot\frac{2}{1} = \frac{1}{2}
$$

We can obtain the last expression also through Bayes' rule:

$$
P(R|D) = \frac{P(D|R)P(R)}{P(D)} = \frac{5}{9}\frac{9}{20}\frac{2}{1} = \frac{1}{2
}
$$

The same works with multiple variables:

$$
P(R,\neg D|M) = \frac{P(R,\neg D, M)}{P(M)} = \frac{7}{80}\frac{2}{1} = \frac{7}{40} 
$$

$$
P(M|D,R) = \frac{P(M,D|R)}{P(D|R)} = \frac{P(M,D,R)}{P(D|R)P(R)}
= \frac{P(M,D,R) P(R)}{P(D,R)P(R)}
= \frac{P(M,D,R)}{P(D,R)} = \frac{9}{40}\frac{4}{1} = \frac{9}{10}
$$

## Conclusions

In frequentist approaches for inference, parameters are adjusted so that they maximize the likelihood function.

In Bayesian statistics we will use the data to obtain probability distributions for all model parameters.
