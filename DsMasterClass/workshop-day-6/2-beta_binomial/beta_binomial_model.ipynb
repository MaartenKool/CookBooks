As a first example, we are going to look at the master example of Bayesian modeling, the beta binomial model. We will use the example from the previous lesson.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=1.5)

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
from scipy import stats
```

We set up a certain number of Bernoulli trials and record how many ones and zeros we obtain.

```python
# define the Bernoulli distribution
p = 0.4
bern = stats.bernoulli(p)
# number of trials
n = 25
# create random samples
sample = bern.rvs(size=n, random_state=2)
# number of ones contained in values
k = sample.sum()
print(sample)
print(np.unique(sample, return_counts=True))
```

Even though our distribution parameter is $p=0.4$, we get only a small fraction of ones in this case.

```python
np.mean(sample)
```

```python
sample.sum()
```

The maximum likelihood estimate would indicate this fraction as the optimal estimate of the distribution parameter.

```python
pvalues = np.linspace(0.01, 0.99, 100)
yvalues = [(sample*np.log(p_i)+(1-sample)*np.log(1-p_i)).sum()
           for p_i in pvalues]
likelihood_max = k/n
yvalues_max = k*np.log(likelihood_max)+(n-k)*np.log(1-likelihood_max)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(pvalues, yvalues)
ax.axhline(yvalues_max, 0, 1)
ax.scatter(likelihood_max, yvalues_max, color='r',
           marker='x', s=100, label='max')
ax.set_title(
    'The log-likelihood function at its maximum for n={}, k={}\n'.format(n, k))
ax.set_xlabel('p')
ax.set_ylabel('log-likelihood')
ax.legend()
plt.show()
```

We have a sequence of Bernoulli trials and observe a certain number of ones and zeros. The maximum likelihood estimate for the probability of having a one gives 0.2 for the obtained data even though we know that the true value is 0.4.

Since we do not care about the order, we can equally well consider the binomial distribution

$$ P(k|n, p) = \binom{n}{k} p^k (1 - p)^{n-k} $$

We can determine how likely it is to see as many ones as we do.

```python
binom = stats.binom(n, p)
binom.pmf(k)
```

The maximum likelihood estimate gives us a point estimate. But what we really would like to determine in Bayesian modeling is how likely $p=0.2$ is given the data. Since we know that the true distribution has $p=0.4$, we might splice that into the prior believe. In any case, the prior distribution should associate some probability with all possible values for the parameter $p$. It makes sense to assume that all values between zero and one are possible, so we would really like to obtain a probability density for the posterior, even though the likelihood function is discrete. We could form now assumptions how the prior might look like, but in any case it has to be a probability density function with support on the range zero to one as well.

A convenient choice is the beta distribution. This is a distribution with at most one maximum with support on the desired range. It will serve for associating probabilities with our probability parameter.

The beta distribution has the form

$$ {\rm pdf}(x|\alpha,\beta) = \frac{\Gamma (\alpha + \beta) }{ \Gamma (\alpha) \Gamma (\beta) }x^{\alpha-1}(1-x)^{\beta-1} $$

where the Gamma function is defined as 

$$ \Gamma(z) =
\int_0^{\infty} x^{z-1} e^{-x} dx 
$$

which gives a finite number for all real numbers $z$ except 0 and negative integers.

For positive integers, the evaluation of the integral gives 

$$\Gamma(z)=(z-1)!$$

In general one has $\Gamma(z+1)=z\,\Gamma(z)$ and $\Gamma(1)=1$.

The beta distribution has two parameters $\alpha$ and $\beta$ which we can freely adjust as long as they are both larger than zero. The mode (for $\alpha,\beta>1$) of the distribution is located at

$$
\frac{\alpha-1}{\alpha+\beta-2}
$$

We can depict the beta distribution for various parameter choices:

```python
params = np.array([0.1, 1, 2, 4, 8])
alpha_m, beta_m = np.meshgrid(params, params)
pvals = np.linspace(0., 1., 100)

fig, ax = plt.subplots(nrows=5, ncols=5, sharex=True,
                       sharey=True, figsize=(12, 12))

for i in range(5):
    for j in range(5):
        beta_dist = stats.beta(alpha_m[i, j], beta_m[i, j])
        beta_vals = beta_dist.pdf(pvals)

        ax[i, j].plot(pvals, beta_vals, lw=3,
                      label=r'$\alpha={}$, $\beta={}$'.format(
                          alpha_m[i, j], beta_m[i, j])+'\n'
                      )
        ax[i, j].set_xticks([0, 0.2, 0.4, 0.6, 0.8, 1])
        ax[i, j].set_xticklabels(
            [0, 0.2, 0.4, 0.6, 0.8, 1], rotation=45, fontsize=8)
        if i == 4:
            ax[i, j].set_xlabel('p', fontsize=16)
        if j == 0:
            ax[i, j].set_ylabel(
                r'pdf(p | $\alpha$, $\beta$)'+'\n', fontsize=16)
        ax[i, j].legend(loc='upper right', fontsize=8)

fig.suptitle('Posterior and prior beta pdf', fontsize=18)

plt.show()
```

## Calculating the posterior


With this prior we are writing the following model:

$$
P(p|n,k;\alpha,\beta) = \frac{P(x|p)P(p;\alpha,\beta)}{\int_0^1 d p P(x|p)P(p;\alpha,\beta)}
$$

To calculate the posterior we have to do the integration in the denominator.

For now however, we discard the normalizing denominator and simply write

$$ P(p|n,k) \propto P(n,k|p) \cdot P(p) $$

Inserting the binomial likelihood and the beta prior gives

$$ 
P(p|n,k) 
\propto
\binom{n}{k} p^k (1 - p)^{n-k} \cdot \frac{\Gamma (\alpha_{\rm prior} + \beta_{\rm prior}) }{ \Gamma (\alpha_{\rm prior}) \Gamma (\beta_{\rm prior}) }p^{\alpha_{\rm prior}-1}(1-p)^{\beta_{\rm prior}-1}
$$

Reordering terms, this is the same as

$$
P(p|n,k) 
\propto 
\binom{n}{k} \,
\frac{\Gamma (\alpha_{\rm prior} + \beta_{\rm prior}) }{ \Gamma (\alpha_{\rm prior}) \Gamma (\beta_{\rm prior}) } \,
p^k (1 - p)^{n-k}\, p^{\alpha_{\rm prior}-1}(1-p)^{\beta_{\rm prior}-1}
$$

Let's now define a constant $c$ as:

$$ c = \binom{n}{k} \cdot 
\frac{\Gamma (\alpha_{\rm prior} + \beta_{\rm prior}) }
{ \Gamma (\alpha_{\rm prior}) \Gamma (\beta_{\rm prior}) } $$

Now our formula for the unnormalized posterior is:

$$ \begin{eqnarray*}
P(p|n,k) &\propto& c \cdot p^k (1 - p)^{n-k} \cdot p^{\alpha_{\rm prior}-1}(1-p)^{\beta_{\rm prior}-1} \\
P(p \;|\; n,k) &\propto& c \cdot p^{k + \alpha_{\rm prior} - 1} (1-p)^{n - k + \beta_{\rm prior} - 1}
\end{eqnarray*}
$$

If we define a new $\alpha$ and $\beta$

$$ \begin{eqnarray*}
\alpha_{\rm posterior} &=&  k+\alpha_{\rm prior} \\
\beta_{\rm posterior} &=& n-k+\beta_{\rm prior}
\end{eqnarray*} $$

the posterior distribution turns out to be a beta distribution itself. This insight fixes the normalization constant $c$.

The properly normalized posterior distribution is then

$$ P(p|n,k)=
\frac{\Gamma(n+\alpha_{\rm prior}+\beta_{\rm prior})}{\Gamma(k+\alpha_{\rm prior})\Gamma(n-k+\beta_{\rm prior})}
p^{k+\alpha_{\rm prior}-1}(1-p)^{n-k+\beta_{\rm prior}-1}
$$

and to obtain this result we could skip doing the integration.

We found a particular situation here: we combined a likelihood function with a prior distribution and obtained a posterior distribution which is of the same type as the prior distribution, even though with different shape parameters. Such pairs of distributions are referred to as conjugate priors. Before the advent of MCMC, Bayesian modeling was mostly limited by knowledge about such conjugate pairs. A nice list of the main ones can be found [here](https://en.wikipedia.org/wiki/Conjugate_prior). In this way we can determine the posterior in a purely algebraic way without solving integrals.

## Let's see what we get for our example

We obtained in 25 trials 5 successes (i.e. ones), even though we know the true $p$ is 0.4.
We can choose the parameters for our prior accordingly.  We can make the prior very flat, telling we have no strong prior belief, or very steep, for example peaking at 0.4.


E.g. with 

$$
\alpha_{\rm prior} = 1, \beta_{\rm prior} = 1 
$$

we get 

$$
\alpha_{\rm posterior} = 1+5 = 6, \beta_{\rm posterior} = 1+20 = 21 
$$

whereas with 


$$
\alpha_{\rm prior} = 40, \beta_{\rm prior} = 60 
$$

we get 

$$
\alpha_{\rm posterior} = 40+5 = 45, \beta_{\rm posterior} = 60+20 = 80 
$$
