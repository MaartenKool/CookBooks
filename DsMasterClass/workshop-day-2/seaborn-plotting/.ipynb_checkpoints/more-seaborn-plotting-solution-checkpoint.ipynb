```python
import warnings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=1.5)

%config InlineBackend.figure_format = 'retina'
%matplotlib inline

warnings.simplefilter('ignore')
```

```python
df = pd.read_csv(
    '../../../dataset-collection/Bike-Sharing-Dataset/hour.csv')
```

```python
df.set_index(pd.to_datetime(df.dteday.astype(str)+'-' +
                            df.hr.map(lambda x: str(x) if x > 9 else '0'+str(x))), inplace=True)
```

```python
df = df.drop(['instant','dteday'],axis=1)
```

```python
df.head()
```

## Produce a heatmap of the correlation matrix

```python
plt.figure(figsize=(8, 8))
sns.heatmap(df.corr(), annot=False)
plt.show()
```

## Produce a scatterplot of windspeed versus temperature color-coded by counts

```python
sns.scatterplot(x='temp', y='windspeed', hue='cnt', data=df, 
                palette='rainbow')
plt.legend(loc=[1.1,0])
plt.show()
```

## Produce box plots of all the variables

Don't forget standardization.

```python
df_stand = df.select_dtypes(include=['int', 'float'])
df_stand = (df_stand-df_stand.mean())/df_stand.std()
sns.boxplot(data=df_stand, orient='h')
plt.show()
```

## Produce a pointplot of temperature or humidity versus season resolved by weathersituation

```python
sns.pointplot(x='season', y='temp', hue='weathersit',
              data=df, orient='v', linewidth=2)
sns.pointplot(x='season', y='temp', data=df, orient='v',
              linewidth=2, linestyle='-.', color='k')
plt.legend(loc=[1.1, 0], title='weathersit', title_fontsize=16)
plt.show()
```

```python
sns.pointplot(x='season', y='hum', hue='weathersit',
              data=df, orient='v', linewidth=2)
sns.pointplot(x='season', y='hum', data=df, orient='v',
              linewidth=2, linestyle='-.', color='k')
plt.legend(loc=[1.1, 0], title='weathersit', title_fontsize=16)
plt.show()
```

## Produce a countplot of weathersituations by season

```python
sns.countplot(x='weathersit', hue='season', data=df,
              orient='v', linewidth=2)
plt.legend(loc=[1.1, 0], title='season', title_fontsize=16)
#plt.yscale('log')
plt.show()
```

## Produce a pair plot with temperature, humidity and counts

```python
mask = ['temp', 'hum', 'cnt']
```

```python
g = sns.PairGrid(df, hue="season", x_vars=mask, y_vars=mask, diag_sharey=False)
g = g.map_diag(sns.distplot, kde_kws={'linewidth': 3})
g = g.map_upper(plt.scatter, alpha=0.5)
g = g.map_lower(sns.kdeplot)
g = g.add_legend()
```
