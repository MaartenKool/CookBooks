# Simpson's paradox

_Author: Christoph Rahmede_

Simpson's paradox is a classical example where probabilistic reasoning seems to fail. Conditioning on additional variables seems to lead to different results than without the conditioning. Looking at aggregated or segmented data seems to lead to contradicting results. The paradox can be resolved if cause and effect are properly taken into account. It may turn out that one indeed has to look at the aggregated data, or at the segmented data, or that one needs further input to come to a sensible conclusion.

## University admissions

A famous example stems from university admission of UC Berkeley from the seventies. It turned out that even though individual departments had taken measures to admit more female students, the university-wide admission rate seemed to indicate a clear gender bias.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=1.5)

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
import statsmodels.api as sm
```

```python
data = sm.datasets.get_rdataset('UCBAdmissions', package='datasets')
```

```python
data.keys()
```

```python
print(data.__doc__)
```

### The data

```python
data.data
```

```python
data.data.shape
```

```python
data.data.columns
```

### Aggregating over frequency segmented by gender and admission

```python
data.data.pivot_table(columns=['Gender','Admit'],
                      values=['Freq'],
                      aggfunc=[np.sum],
                      margins=True)
```

### Aggregating over frequency segmented by gender, admission and department

```python
pivot_table = data.data.pivot_table(
                index='Dept',
                columns=['Gender','Admit'],
                values='Freq',
                aggfunc=np.sum,
                margins=True)
pivot_table
```

### Create columns for number of applicants and admission rates

```python
pivot_table[('Female','Applicants')] = pivot_table.apply(lambda x: (x[0]+x[1]),axis=1)
pivot_table[('Female','Admit_rate')] = pivot_table.apply(lambda x: x[0]/(x[0]+x[1]),axis=1)
pivot_table[('Male','Applicants')] = pivot_table.apply(lambda x: (x[2]+x[3]),axis=1)
pivot_table[('Male','Admit_rate')] = pivot_table.apply(lambda x: x[2]/(x[2]+x[3]),axis=1)
```

```python
pivot_table
```

```python
pivot_table[[('Female','Admit_rate'),('Male','Admit_rate')]].plot(kind='barh')
plt.legend(loc=[1.1,0])
plt.show()
```

### Illustrating the results

We can illustrate what happens by first looking only at two departments. In both cases, female students have higher admission rates, but in the combined group the admission rate for male students is higher. This is due to the higher number of applicants of the latter which can be seen as the result of vector addition for vectors having as components the number of applicants and the number of admissions. 

In the plot, dashed lines indicate female and full lines male proportions.

```python
from matplotlib.colors import Normalize
import matplotlib.cm as cm
from matplotlib.colors import ListedColormap

norm = Normalize(vmin=1, vmax=len(pivot_table))
cmap = ListedColormap(sns.color_palette("husl", len(pivot_table)))
```

```python
fig, ax = plt.subplots(figsize=(12,12))

# variables for arrow layout
hw = 20
hl = 50

row_set = ['A','D']
for i, dep in enumerate(row_set):
    

    ax.arrow(0, 0, 
             pivot_table[('Male','Applicants')][dep], 
             pivot_table[('Male','Admitted')][dep], 
             fc=cmap(i), ec=cmap(i), head_width=hw, head_length=hl, 
             length_includes_head=True,label='x',lw=3,linestyle='-',alpha=0.5)
            
    ax.arrow(0, 0, 
             pivot_table[('Female','Applicants')][dep], 
             pivot_table[('Female','Admitted')][dep], 
             fc=cmap(i), ec=cmap(i), head_width=hw, head_length=hl, 
             length_includes_head=True,label='x',lw=3,linestyle='--',alpha=0.5)


ax.arrow(0, 0, 
             pivot_table.loc[row_set][('Male','Applicants')].sum(), 
             pivot_table.loc[row_set][('Male','Admitted')].sum(), 
             fc='k', ec='k', head_width=hw, head_length=hl, 
             length_includes_head=True,label='x',lw=3,linestyle='-',alpha=0.5)
            
ax.arrow(0, 0, 
             pivot_table.loc[row_set][('Female','Applicants')].sum(), 
             pivot_table.loc[row_set][('Female','Admitted')].sum(), 
             fc='k', ec='k', head_width=hw, head_length=hl, 
             length_includes_head=True,label='x',lw=3,linestyle='--',alpha=0.5)
    
ax.set_xlim([0,800])
ax.set_ylim([0,800])
ax.set_xlabel('Number of applicants')
ax.set_ylabel('Number of admissions')

plt.show()
```

In the same way, we can illustrate the proportions for the individual groups as vectors and the combined group vector will result from vector addition of the group vectors.

```python
fig, ax = plt.subplots(figsize=(12,12))
hw = 20
hl = 50
for i, dep in enumerate(pivot_table.index):
    
    ax.set_label(pivot_table.index)
    ax.arrow(0, 0, 
             pivot_table[('Male','Applicants')][dep], 
             pivot_table[('Male','Admitted')][dep], 
             fc=cmap(i), ec=cmap(i), head_width=hw, head_length=hl, 
             length_includes_head=True,label='x',lw=3,linestyle='-',alpha=0.5)
            
    ax.arrow(0, 0, 
             pivot_table[('Female','Applicants')][dep], 
             pivot_table[('Female','Admitted')][dep], 
             fc=cmap(i), ec=cmap(i), head_width=hw, head_length=hl, 
             length_includes_head=True,label='x',lw=3,linestyle='--',alpha=0.5)
    
ax.set_xlim([0,800])
ax.set_ylim([0,800])
ax.set_xlabel('Number of applicants')
ax.set_ylabel('Number of admissions')

plt.show()
```

## Example

Let's consider a typical example due to [Judea Pearl](#Pearl).

Consider a medical trial on two groups, 40 male and 40 female patients.
Does the drug lead to increased recovery rate? Should the drug be recommended?

|-|Recovered|Not recovered|Recovery rate|
|----|----|----|----|
|Males|
|Given drug|18|12|0.6|
|Not given drug|7|3|0.7|
|Females|
|Given drug|2|8|0.2|
|Not given drug|9|21|0.3|
|Combined|
|Given drug|20|20|0.5|
|Not given drug|16|24|0.4|

In this case, gender is a confounding variable since it has an impact on both, the drug being taken and the recovery rate.

We have the following situation stated by conditional probabilities where either we do not condition on gender and look at the combined groups, or we condition on gender and look at the respective groups:

$$
\begin{eqnarray}
P(R|D) = 0.5 &>& P(R|\neg D) = 0.4 \\
P(R|D, M) = 0.6 &<& P(R|\neg D, M) = 0.7 \\
P(R|D, F) = 0.2 &<& P(R|\neg D, F) = 0.3 \\
\end{eqnarray}
$$

Here $\neg D$ stands for not being given the drug. We clearly see the reversal of the relationship.

Note that we also have a higher recovery rate among men and a higher drug use among men:

$$
\begin{eqnarray}
P(R|M) = 0.625 &>& P(R|F) = 0.275 \\
P(D|M) = 0.75 &>& P(D|F) = 0.25
\end{eqnarray}
$$


So we can say that a male patient recovers faster and is more likely to take the drug. Both, male and female patients have a higher recovery rate without the drug. Apparently, it is not simply the conditional probability $P(R|D)$ for the combined group that we are after.

Rather we would like to obtain the average of the recovery rate across the two groups weighted by the probability for being in any of the two gender groups:

$$
P(R|D,M)P(M)+P(R|D,F)P(F) = 0.6\cdot 0.5 + 0.2\cdot 0.5 = 0.4
$$

whereas 

$$
P(R|\neg D,M)P(M)+P(R|\neg D,F)P(F) = 0.7\cdot 0.5 + 0.3\cdot 0.5 = 0.5
$$

We cannot arrive at the conclusion by purely looking at conditional probabilities. Instead, a careful analysis about which variables can have a causal effect on others can give the correct answer.
