## Variable types

In any kind of dataset we will meet variables of different type.

- Categorical variables
    - Nominal: categories without order, e.g. male/female
    - Ordinal: categories with order, e.g. ranking from a race
    
- Metric variables
    - Discrete, e.g. counts
    - Continuous, e.g. temperature or time

### Bar plot

Bar plots serve for comparing metric variables across different categories.

````python
counts = df.origin.value_counts()
plt.bar(counts.index, counts.values, color=['r', 'b', 'g'])
plt.xlabel('origin')
plt.ylabel('counts')
plt.show()
````

### Getting the brands

We have one entry for each type of car, but many of them belong to the same brand. Let's extract the brand name.

````python
df.name.map(lambda x: x.split(' ')[0]).unique()
````

Since there are some spelling mistakes, let's clean the names up.

````python
df['brand'] = df.name.map(lambda x: x.split(' ')[0]
                          ).replace(
    {'toyouta': 'toyota',
     'maxda': 'mazda',
     'chevroelt': 'chevrolet',
     'vw': 'volkswagen',
     'vokswagen': 'volkswagen',
     'mercedes-benz': 'mercedes'})
````

````python
counts = df.brand.value_counts()
plt.figure(figsize=(12, 12))
plt.barh(counts.index, counts.values)
plt.xlabel('counts')
plt.ylabel('brands')
plt.show()
````

#### Ex: Looking at the different brands, find out what the origin stands for.

### Pie chart

Pie charts are popular, but really only useful if the different parts sum up to 1, and if they are depicted together with the fractions or percentages. With the pure eye it is challenging to compare the exact proportions between different slices. Bar plots are always preferred!

````python
df['continent'] = df.origin.map(
    lambda x: 'America' if x == 1 else 'Europe' if x == 2 else 'Asia')
````

````python
counts = df.continent.value_counts(normalize=True)
plt.pie(counts.values, labels=counts.index, labeldistance=1.1, autopct='%1.1f%%',
        textprops={'fontsize': 20}, explode=[0, 0, 0.1], shadow=True, startangle=90)
plt.show()
````

### Line plot

Also a line plot can serve for depicting a relation between two variables. Keep in mind that to get something useful the x-values should be ordered.

````python
mpgs = []
years = np.sort(df.year.unique())
for year in years:
    mpgs.append(df[df.year == year].mpg.mean())
````

````python
plt.scatter(years, mpgs)
plt.plot(years, mpgs, lw=2)
plt.xticks(years)
plt.xlabel('year')
plt.ylabel('mean mpg')
plt.show()
````

### Boxplot

A box plot is another way for depicting the distribution of a variable. It consists of the following parts (to be read for vertical box plots):

- box: height defined by the interquartile range of the variable (25th and 75th percentile)
- median: the line in the center marking the 50th percentile.
- the whiskers: the vertical lines departing from the box, extending up to the maximal/minimal value or at most up to 1.5 times the interquartile range from the box boundary (can be adjusted by the `whis` argument, 1.5 is the default value). The actual boundary is at the last observation within the whisker distance.
- caps: horizontal lines at the end of the whiskers

Any points outside the range of the whiskers will be represented by individual markers. That serves to discover outliers easily.

Let's look at two examples.

````python
plt.boxplot(df.weight, sym='d', showfliers=True, whis=1.5)
plt.xticks([1], ['weight'])
plt.show()
````

````python
plt.boxplot(df.horsepower, sym='d', showfliers=True, whis=1.5)
plt.xticks([1], ['horsepower'])
plt.show()
````

Let's illustrate what the individual parts stand for.

````python
x_min = 0.9,
x_max = 1.1
median = df.horsepower.median()
h_max = df.horsepower.max()
h_min = df.horsepower.min()
percentile_25 = df.horsepower.describe()['25%']
percentile_75 = df.horsepower.describe()['75%']

upper_whis = percentile_75+1.5*(percentile_75-percentile_25)
lower_whis = percentile_25-1.5*(percentile_75-percentile_25)
upper_cap = df.horsepower[df.horsepower < upper_whis].max()

# that is the upper cap, every point beyond will be represented by a marker
upper_cap
````

````python
plt.boxplot(df.horsepower, sym='d', whis=1.5)
plt.xticks([1], ['horsepower'])
plt.hlines(median, x_min, x_max, color='b')
plt.hlines(percentile_25, x_min, x_max, color='b')
plt.hlines(percentile_75, x_min, x_max, color='b')
plt.hlines(h_min, x_min, x_max, color='b')
plt.hlines(h_max, x_min, x_max, color='b')
plt.hlines(upper_cap, x_min, x_max, color='b')
plt.hlines(upper_whis, x_min, x_max, color='b')

fontsize = 12
arrowprops = dict(arrowstyle='->', color='black', linewidth=1.5)

plt.annotate('median',
             (1.1, median),
             xytext=(1.2, median),
             fontsize=fontsize,
             arrowprops=arrowprops)

plt.annotate('25th percentile',
             (1.1, percentile_25),
             xytext=(1.2, percentile_25),
             fontsize=fontsize,
             arrowprops=arrowprops)

plt.annotate('75th percentile',
             (1.1, percentile_75),
             xytext=(1.2, percentile_75),
             fontsize=fontsize,
             arrowprops=arrowprops)

plt.annotate(r'$75{\rm th}\,{\rm percentile}\,$'+'\n'+r'$+ 1.5\cdot {\rm IQR}$',
             (1.1, upper_whis),
             xytext=(1.2, upper_whis),
             fontsize=fontsize,
             arrowprops=arrowprops
             )

plt.annotate('upper cap',
             (0.9, upper_cap),
             xytext=(0.7, upper_cap),
             fontsize=12,
             arrowprops=arrowprops
             )

plt.annotate('maximum',
             (1.1, h_max),
             xytext=(1.2, h_max),
             fontsize=12,
             arrowprops=arrowprops
             )

plt.annotate('minimum',
             (1.1, h_min),
             xytext=(1.2, h_min),
             fontsize=12,
             arrowprops=arrowprops
             )

plt.show()
````

We can also plot several variables with boxplots together.

````python
plt.boxplot([df.horsepower, df.mpg, df.weight], sym='d', showfliers=True)
plt.xticks(range(1, 4), ['horsepower', 'mpg', 'weight'])
plt.show()
````

Apparently we have a problem with differing scales. As a remedy, the most common approach is to standardize the variables by subtracting the mean and dividing by the standard deviation. This produces unit-less numbers which express how many standard deviations any observation is above or below the mean. 

````python
def standardizer(x):
    return (x-x.mean())/x.std()
````

````python
plt.boxplot([standardizer(df.horsepower),
             standardizer(df.mpg),
             standardizer(df.weight)],
            sym='d',
            showfliers=True)
plt.xticks(range(1, 4), ['horsepower', 'mpg', 'weight'])
plt.show()
````

### Figure objects

So far we have been working in an interactive way. That is nice for experimenting and small amounts of data since we can easily run a cell again after we changed any aspect of the figure. The more sophisticated way is to work with figure objects which can be saved and whose properties we can modify without having to recreate the whole figure again.

Each matplotlib figure consists of two objects, conventionally referred to as `ax` and `fig`.
The `ax` object contains all the metadata about the figure, `fig` is the figure itself.

````python
fig, ax = plt.subplots()
ax.scatter(years, mpgs)
ax.plot(years, mpgs, lw=2)
ax.set_xticks(years)
ax.set_xlabel('year')
ax.set_ylabel('mean mpg')
plt.show()
````

````python
# the ax object
ax
````

We can now modify attributes of the ax-object and display the updated figure.

````python
ax.set_title('Mean mpg per year')
fig
````

## Combining multiple figures

We can easily create combinations of different figures by specifying how many figures we intend to construct and then referring to each entry with numpy matrix notation. We indicate how many figures we want with the keywords `nrows` and `ncols` in `plt.subplots`. Then we can set up for example the upper left figure with

```python
fig, ax = plt.subplots(nrows=2, ncols=2,figsize=(6,6))
ax[0, 0].scatter(....)
```

## Plotting styles

Matplotlib works with a lot of default parameters. They are stored in a dictionary which you can have a look at with 

```python
plt.rcParams
```

You can change any of those, for example the standard linewidth. Be careful!

````python
standardlinewidth = plt.rcParams['lines.linewidth']
standardlinewidth
````

````python
plt.rcParams['lines.linewidth'] = 8
````

````python
plt.scatter(years, mpgs)
plt.plot(years, mpgs)
plt.xticks(years)
plt.xlabel('year')
plt.ylabel('mean mpg')
plt.show()
````

````python
plt.rcParams['lines.linewidth'] = standardlinewidth
````

````python
plt.scatter(years, mpgs)
plt.plot(years, mpgs)
plt.xticks(years)
plt.xlabel('year')
plt.ylabel('mean mpg')
plt.show()
````

Matplotlib has many different style choices available:

````python
print(plt.style.available)
````

You can pick any of these styles with something like 

```python
plt.style.use('seaborn-dark')
```

That will make this style permanent (for your session).
If you just want to change the style temporarily, try the following:

````python
with plt.style.context(('fivethirtyeight')):
    fig_1, ax_1 = plt.subplots()
    ax_1.scatter(years, mpgs)
    ax_1.plot(years, mpgs, lw=2)
    ax_1.set_xticks(years)
    ax_1.set_xlabel('year')
    ax_1.set_ylabel('mean mpg')
````

````python
with plt.xkcd(randomness=10):
    fig_1, ax_1 = plt.subplots(ncols=2,figsize=(12,4))
    ax_1[0].scatter(years, mpgs, color='r', s=80)
    ax_1[0].plot(years, mpgs, lw=2)
    ax_1[0].set_xticks(years)
    ax_1[0].set_xlabel('year')
    ax_1[0].set_ylabel('mean mpg')
    ax_1[1].boxplot(df.weight, sym='d', showfliers=True, whis=1.5)
    ax_1[1].set_xticks([1], ['weight']);
````

The global style is still the same:

````python
plt.plot(years, mpgs, alpha=0.8)
plt.scatter(years, mpgs,color='r',s=50)
plt.xticks(years)
plt.xlabel('year')
plt.ylabel('mean mpg')
plt.show()
````
