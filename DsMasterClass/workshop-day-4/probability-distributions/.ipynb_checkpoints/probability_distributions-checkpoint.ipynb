## Discrete probability distributions

For discrete distributions, the probabilities of all the possible events are given by the probability mass function (pmf).

### The Discrete Uniform Distribution

In the [discrete uniform distribution](https://docs.scipy.org/doc/scipy/reference/tutorial/stats/discrete_randint.html), we have $n$ possible outcomes $X_i$ which each have probability $1/n$:

$$
{\rm pmf}(X_i;n) = \frac{1}{n}\, , \, i\in \{1,\ldots,n\}
$$

### The Bernoulli distribution

The [Bernoulli distribution](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.bernoulli.html#scipy.stats.bernoulli) describes a process with two possible outcomes, of which one has probability $p$ and the other one has probability $1-p$:

$$  
\begin{eqnarray}
{\rm pmf}(X=1;p)&=& p\\
{\rm pmf}(X=0;p)&=& (1-p)
\end{eqnarray}
$$


### The Binomial distribution

Generalizing the Bernoulli distribution to $n$ trials and asking only for the occurrence $k$ of one of the binary outcomes (let's say 1) irrespective of the order, gives the [binomial distribution](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.binom.html#scipy.stats.binom):

$$ {\rm pmf}(k;n,p) = \binom n k  p^k(1-p)^{n-k} $$


The [binomial coefficient](https://docs.scipy.org/doc/scipy/reference/generated/scipy.special.comb.html#scipy.special.comb) is given by

$$
\binom n k = \frac{n!}{k!(n-k)!}
$$

### Poisson distribution

A discrete distribution allowing for an arbitrary number of observed events (in principle) is given by the [Poisson distribution](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.poisson.html#scipy.stats.poisson):

$${\rm pmf}(k;\lambda) = \frac{\lambda^k e^{-\lambda}}{k!} $$

### Multinomial distribution

 
The [multinomial distribution](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.multinomial.html#scipy.stats.multinomial) generalizes the binomial distribution to more than two possible outcomes.

### Implementation in scipy

```python
discrete.rvs(10)
```

#### Illustration

```python
x = np.arange(1, 5)
discrete = stats.randint(1, 5)

fig, ax = plt.subplots(ncols=2, figsize=(16, 6))

ax[0].bar(x, discrete.pmf(x))
ax[0].set_xlabel('Event')
ax[0].set_ylabel('PMF')

ax[1].bar(x, discrete.cdf(x))
ax[1].set_xlabel('Event')
ax[1].set_ylabel('CDF')

fig.suptitle('Discrete Uniform Distribution, p = {:.3f}'.format(
    1/len(x)), fontsize=24)
plt.show()
```

```python
p = 0.3
x = np.arange(0, 2)
discrete = stats.bernoulli(p)

fig, ax = plt.subplots(ncols=2, figsize=(16, 6), sharey=True)

ax[0].bar(x, discrete.pmf(x))
ax[0].set_xlabel('Event')
ax[0].set_ylabel('PMF')
ax[0].set_xticks([0, 1])

ax[1].bar(x, discrete.cdf(x))
ax[1].set_xlabel('Event')
ax[1].set_ylabel('CDF')
ax[1].set_xticks([0, 1])

fig.suptitle('Bernoulli Distribution, p = {:.3f}'.format(p), fontsize=24)
plt.show()
```

```python
p = 0.3
n = 20
x = np.arange(0, n)
discrete = stats.binom(n, p)

fig, ax = plt.subplots(ncols=2, figsize=(16, 6), sharey=False)

ax[0].bar(x, discrete.pmf(x))
ax[0].set_xlabel('Event')
ax[0].set_ylabel('PMF')

ax[1].bar(x, discrete.cdf(x))
ax[1].set_xlabel('Event')
ax[1].set_ylabel('CDF')

fig.suptitle(
    'Binomial Distribution, n = {}, p = {:.3f}'.format(n, p), fontsize=24)
plt.show()
```

```python
lambda_ = 3
n = 20
x = np.arange(0, n)
discrete = stats.poisson(lambda_)

fig, ax = plt.subplots(ncols=2, figsize=(16, 6), sharey=False)

ax[0].bar(x, discrete.pmf(x))
ax[0].set_xlabel('Event')
ax[0].set_ylabel('PMF')

ax[1].bar(x, discrete.cdf(x))
ax[1].set_xlabel('Event')
ax[1].set_ylabel('CDF')

fig.suptitle('Poisson Distribution, p = {:.3f}'.format(lambda_), fontsize=24)
plt.show()
```

## Continuous probability distributions

For continuous distributions, the probabilities of possible events are obtained from the probability density function (pdf). The pdf however does not indicate probabilities directly. Probabilities are obtained for the continuous parameter being in a certain range as the area (or volume in the higher-dimensional case) under the probability density function.

### Continuous uniform distribution

For the [uniform distribution](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.uniform.html#scipy.stats.uniform), the probability density is the same in a finite range:

$$
{\rm pdf}(X; a,b) = \frac{1}{b-a} \, \ \text{for X in [a,b], else 0} 
$$

### Normal distribution

For the [normal distribution](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html#scipy.stats.norm), the probability density function is

$$ {\rm pdf}(X; \mu, \sigma)
= \frac{1}{\sqrt{2\pi\sigma^2} } \; e^{ -\frac{(X-\mu)^2}{2\sigma^2} } $$

Here, $\mu$ is the mean of the distribution and $\sigma$ is the standard deviation of the distribution.

### Exponential distribution

The continuous [exponential distribution](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html#scipy.stats.expon) is defined as

$$ {\rm pdf}(x;\lambda) = \lambda  e^{-\lambda x} $$

The random variable $x$ can only take positive values, $\lambda$ is the shape parameter of the distribution. The distribution function is monotonically decreasing.

### Pareto distribution

The [Pareto distribution](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html#scipy.stats.expon) (also known as power-law distribution) is given by

$$
{\rm pdf}(x;b) = \frac{b}{x^{b+1}}
$$

with $x\ge 1$ and $b>0$. The variable $x$ measures the ratio between any value and the lowest possible value, hence it cannot become smaller than 1.

This distribution decays much slower towards the tail than the normal or exponential distribution, it has "fat tails".

For this distribution higher moments do not exist. For example, if $b\le 2$, the variance does not exist, and if $b\le 1$ both mean and variance do not exist.

### Illustration

The set-up works in quite the same way, only that now we call the probability density function rather than the probability mass function.

```python
left = 0
right = 0.5
continuous = stats.uniform(left, right)
```

```python
x = np.linspace(0, 0.5, 10)
continuous.pdf(x)
```

```python
left = 0
right = 0.5
x = np.linspace(left, right, 10)
continuous = stats.uniform(left, right)

fig, ax = plt.subplots(ncols=2, figsize=(16, 6), sharey=False)
ax[0].plot(x, continuous.pdf(x), lw=4)
ax[0].set_xlabel('x')
ax[0].set_ylabel('PDF')
ax[0].set_ylim([0, 2.5])

ax[1].plot(x, continuous.cdf(x), lw=4)
ax[1].set_xlabel('x')
ax[1].set_ylabel('CDF')

fig.suptitle('Continuous Uniform Distribution, a = {}, b = {}'.format(
    left, right), fontsize=24)
plt.show()
```

```python
mu = 0
sigma = 1
x = np.linspace(mu-3*sigma, mu+3*sigma, 10000)
continuous = stats.norm(mu, sigma)

fig, ax = plt.subplots(ncols=2, figsize=(16, 6), sharey=False)
ax[0].plot(x, continuous.pdf(x), lw=4)
ax[0].set_xlabel('x')
ax[0].set_ylabel('PDF')

ax[1].plot(x, continuous.cdf(x), lw=4)
ax[1].set_xlabel('x')
ax[1].set_ylabel('CDF')

fig.suptitle('Normal Distribution, mu = {}, sigma = {}'.format(
    mu, sigma), fontsize=24)
plt.show()
```

```python
lambda_ = 1
x = np.linspace(0, 10, 1000)
continuous = stats.expon(scale=lambda_)

log_scale = False

fig, ax = plt.subplots(ncols=2, figsize=(16, 6), sharey=False)
ax[0].plot(x, continuous.pdf(x), lw=4)
ax[0].set_xlabel('x')
ax[0].set_ylabel('PDF')

ax[1].plot(x, continuous.cdf(x), lw=4)
ax[1].set_xlabel('x')
ax[1].set_ylabel('CDF')

if log_scale == True:
    ax[0].set_yscale('log')
    ax[1].set_yscale('log')

fig.suptitle('Exponential Distribution, lambda = {}'.format(
    lambda_), fontsize=24)
plt.show()
```

```python
b = 1.5
x = np.linspace(0, 20, 1000)
continuous = stats.pareto(b)

log_scale = False

fig, ax = plt.subplots(ncols=2, figsize=(16, 6), sharey=False)
ax[0].plot(x, continuous.pdf(x), lw=4)
ax[0].set_xlabel('x')
ax[0].set_ylabel('PDF')
ax[0].set_xlim([1, ax[0].get_xlim()[1]])

ax[1].plot(x, continuous.cdf(x), lw=4)
ax[1].set_xlabel('x')
ax[1].set_ylabel('CDF')
ax[1].set_xlim([1, ax[1].get_xlim()[1]])

if log_scale == True:
    ax[0].set_xscale('log')
    ax[1].set_xscale('log')
    ax[0].set_yscale('log')
    ax[1].set_yscale('log')


fig.suptitle('Pareto Distribution, b = {}'.format(b), fontsize=24)
plt.show()
```

## Other interesting continuous distributions

### The student/t-distribution

The [t-distribution](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.t.html#scipy.stats.t) occurs in the context of confidence intervals around estimated means and in hypothesis testing related to means of two samples being statistically significantly different.

### The chi-squared ($\chi^2$) distribution

The [$\chi^2$ distribution](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.chi2.html#scipy.stats.chi2) occurs for example in the context of hypothesis testing (testing for equality of standard deviations of two samples).

### The beta distribution

The [beta distribution](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.beta.html#scipy.stats.beta) is a continuous distribution for random variables limited to values between 0 and 1. We will meet it in the context of Bayesian modeling. We will need it to model the probability of having a certain probability for Bernoulli trials.

## Common notation

To express that a continuous random variable follows a normal distribution with mean $\mu$ and standard deviation $\sigma$ one often writes

$$
X \sim N(\mu,\sigma)
$$

Similarly for a discrete random variable following a Poisson distribution with rate $\lambda$ one would write

$$
X \sim {\rm Poisson}(\lambda)
$$
