- Setup two different Pareto distributions, one with b=1.5 and one with b=0.5. 
- Create a certain amount of random variate samples from each of the distributions and determine the mean.
- Repeat that process (e.g. a 100 or 1000 times), determine the mean of all the means and plot a histogram of all the means.
- Once done, repeat the whole process. How do your histograms change?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=1.5)

%config InlineBackend.figure_format = 'retina'
%matplotlib inline

from scipy import stats
```

```python
n = 1000
means_05 = [stats.pareto(0.5).rvs(100).mean() for i in range(n)]
means_15 = [stats.pareto(1.5).rvs(100).mean() for i in range(n)]
print(np.mean(means_05),np.mean(means_15))
```

```python
plt.hist(means_05,label='b=0.5',bins=np.logspace(0,9,50))
plt.hist(means_15,label='b=1.5',bins=np.logspace(0,9,50))
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.show()
```
