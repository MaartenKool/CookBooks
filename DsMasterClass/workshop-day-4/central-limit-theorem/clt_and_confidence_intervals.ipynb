## Sample mean

When obtaining a sample of data, we can calculate the mean of that sample. If the data in the sample has been collected independently from the same population, we can make estimates about the probability of obtaining a certain sample mean based on the population mean.
  
For $X_1, \ldots, X_n$ independent identically distributed random variables, the sample mean is given by

$$
\bar{X} = \frac{X_1+X_2+\ldots+X_n}{n}
$$

## The central limit theorem

1. For a population with mean $\mu$ the mean of the sampling distribution of means is 

$$
E(\bar{X}) = \mu_{\bar{X}} = \mu
$$

2. If samples are drawn randomly and with replacement from a population with mean $\mu$ and standard deviation $\sigma$ the variance of the sampling distribution of means is given by

$$
E\left[(\bar{X}-\mu)^2\right] = \sigma_{\bar{X}}^2 = \frac{\sigma^2}{n}
$$

3. If the population is normally distributed with mean $\mu$ and standard deviation $\sigma$, the sample means are normally distributed with mean $\mu$ and variance $\sigma^2/n$.

4. If the population follows some arbitrary distribution with finite mean $\mu$ and finite standard deviation $\sigma$, then the standardized random variable resulting from the sample mean 

$$
Z = \frac{\bar{X}-\mu}{\sigma/\sqrt{n}}
$$

is asymptotically normally distributed with 

$$
\lim_{n\rightarrow\infty}P(Z\le z) 
= \frac{1}{\sqrt{2\pi}}\int_{-\infty}^z du\ e^{-\frac{u^2}{2}}
$$

Formulation of the theorem from 

M. R. Spiegel, J. Schiller, R. A. Srinivasan: Probability and Statistics (4th edition), Schaum's outlines, McGraw Hill, 2013

## Data demonstration


Let's again look at our dataset containing many variables with very diverse distributions. We are going to demonstrate that the central limit theorem hold by sampling a certain amount of observations from the data with replacement, calculating the mean each time and plotting the distribution of the means.

```python
df = pd.read_csv(
    '../../../dataset-collection/Bike-Sharing-Dataset/hour.csv')
df.drop(['instant', 'dteday'], axis=1, inplace=True)
df.head()
```

```python
fig, ax = plt.subplots(figsize=(12, 12))
df.hist(density=True, ax=ax)
fig.tight_layout()
plt.show()
```

Let's look at just one of the variables.

```python
df[['temp']].hist(density=True)
plt.show()
```

Now we sample from  that variable an amount of $n$ observations, determine the mean, and repeat that process many times.

```python
n = 100
col = 'temp'
sample_means = np.array([df[col].sample(n, replace=True).mean()
                         for i in range(10000)])

mean, standard = df[col].mean(), df[col].std()/n**0.5
x = np.linspace(mean-3*standard, mean+3*standard)
norm = stats.norm(loc=df[col].mean(), scale=df[col].std()/n**0.5)
```

Compare the histogram with the normal distribution which should describe the distribution of means according to the central limit theorem.

```python
plt.hist(sample_means, bins=50, density=True)
plt.plot(x, norm.pdf(x))
plt.show()
```

That looks roughly normal. A better visualization is given through a quantile-quantile plot, for example implemented in statsmodels.

```python
import statsmodels.api as sm
```

```python
sm.graphics.qqplot(sample_means, line='45', fit=True)
plt.show()
```

## Confidence Interval Estimates of Population Parameters

If we know the distribution of a random variable we can say how likely it is to find a random variable within a certain range. The same is true for any sampling statistic of which we know the distribution. 

In particular we know that under very general conditions sample means are (nearly) normally distributed with mean and standard deviation set by the population mean, standard deviation and sample size.

If the population variance is known, we can use the sample estimate of the mean to establish a confidence interval

$$
\bar{X}\pm z_{\alpha/2}\frac{\sigma}{\sqrt{n}}
$$

with critical values $z_{\alpha/2}$ determined by the level of confidence which should be reached. The critical values correspond to percentile boundaries of the normal distribution.

Due to the central limit theorem, we do not have to know exactly how $X$ itself is distributed, but we have to require large enough sample sizes.

If the population itself is normally distributed we can use the exact result which even holds for small sample sizes:

$$
\bar{X}\pm t_{\alpha/2}\frac{\sigma}{\sqrt{n}}
$$

If the population is not normally distributed and the population standard deviation is not known, i.e. it has to be estimated from the sample as sample standard deviation $S$, then for large enough sample size one can use 

$$
\bar{X}\pm t_{\alpha/2}\frac{S}{\sqrt{n}}
$$

## Interpretation of confidence interval

A 95 % confidence interval can be interpreted as the interval which would contain the true population parameter of a probability distribution in 95 % of the cases of repeated sampling.

```python
norm = stats.norm(loc=0, scale=1)
```

```python
norm.interval(0.99)
```

```python
z_95 = norm.ppf(1-0.05/2)
z_95
```

```python
n = 100
standard = 1
intervals = [np.mean(norm.rvs(n))+z_95*standard/n**0.5 *
             np.array([-1, 1]) for i in range(1000)]
```

```python
np.mean([1 if np.allclose(np.sign(interval), np.array([-1, 1]))
         else 0 for interval in intervals])
```

```python
plt.figure(figsize=(12, 12))
for i in range(len(intervals[:101])):
    color = 'k'
    if (intervals[i][0] > 0) or (intervals[i][1] < 0):
        color = 'r'
        print('iteration not covering true value:', i)
    plt.hlines(i, intervals[i][0], intervals[i][1], color=color, alpha=0.5)
plt.vlines(0, 0, i, color='k')
plt.show()
```

### Confidence interval for the variance

In other cases than means, we usually need to be informed about the distribution of the population. In the case of a normally distributed population (like the sample means) we can for example say

$$
\chi^2_{0.025}\le\frac{(n-1)S^2}{\sigma^2}\le \chi^2_{0.975}
$$

and therefore we obtain the estimate

$$
\frac{\sqrt{n-1}S}{\chi_{0.975}}\le \sigma \le \frac{\sqrt{n-1}S}{\chi_{0.025}}
$$

For large $n$, the $\chi^2$-distribution  is well-approximated by a normal distribution.

Similarly for variance ratios (from normally distributed populations) one can obtain

$$
\frac{1}{F_{0.95}}\frac{S_1^2}{S_2^2}
\le \frac{\sigma_1^2}{\sigma_2^2}
\le \frac{1}{F_{0.05}}\frac{S_1^2}{S_2^2}
$$

with the Fisher-distribution $F$.

```python
n = 100
chi_2 = stats.chi2(n-1)
chi2_bound = np.array(chi_2.interval(0.95))
chi2_bound
```

```python
x = np.linspace(0, 200, 10000)
y = chi_2.pdf(x)
z = stats.norm(chi_2.mean(),chi_2.std()).pdf(x)
plt.plot(x, y, lw=2, label='$\chi^2$')
plt.plot(x, z, lw=2, label='normal')
plt.legend()
plt.show()
```

```python
n = 100
standard = 1
chi_2 = stats.chi2(n-1)
chi2_bound = np.array(chi_2.interval(0.95))
intervals = [(np.sqrt((n-1)/chi2_bound)*np.std(norm.rvs(n)))
             for i in range(1000)]
```

```python
np.mean([1 if np.allclose(np.sign(interval-1), np.array([1, -1]))
         else 0 for interval in intervals])
```
