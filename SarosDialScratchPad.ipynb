```python
# """
#  Spiral.py is a program which make a polar plot of r = theta^2
#    i.e. the Galilean spiral

#  Paul Eugenio
#  Florida State University
#  PHZ4151C
#  29 Jan 2019
# """

from __future__ import division, print_function
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
# use latex 
plt.rc('text', usetex=False)

#  define parameters
thetaMin, thetaMax = 0, 10*np.pi
steps = 500


# use numpy vectorization to generate plotting values
theta = np.linspace(thetaMin, thetaMax, steps)
#circle
r = np.pi
x = r*np.cos(theta)
y = r*np.sin(-theta)

#Archemedean Spiral
rArch= theta
xArch = rArch*np.cos(theta)
yArch = rArch*np.sin(-theta)

#Galilean Spiral
rGal= theta * theta 
xGal = rGal * np.cos(theta)
yGal = rGal * np.sin(-theta)

#Fermat Spiral
rFerm= np.sqrt(theta)
xFerm = rFerm*np.cos(theta)
yFerm = rFerm*np.sin(-theta)

#Lituus Spiral
rLituus = 1/rFerm
xLituus = rLituus*np.cos(theta)
yLituus = rLituus*np.sin(-theta)


fig, axs = plt.subplots(2, 5, figsize=(20,8))
axs[0, 0].plot([1,2,3,4,5],[r,r,r,r,r])
axs[0, 0].set_title('Axis [0,0]')
axs[1, 0].plot(x, y, 'tab:green')
axs[1, 0].set_title('Plain Circle')
axs[0, 1].plot(rArch, 'tab:orange')
axs[0, 1].set_title('Axis [0,1]')
axs[1, 1].plot(xArch, yArch, 'tab:red')
axs[1, 1].set_title('Axis [1,1]')
axs[0, 2].plot(rGal, 'tab:green')
axs[0, 2].set_title('Axis [1,0]')
axs[1, 2].plot(xGal,yGal, 'tab:green')
axs[1, 2].set_title('Axis [1,0]')
axs[0, 3].plot(rFerm, 'tab:green')
axs[0, 3].set_title('Axis [1,0]')
axs[1, 3].plot(xFerm,yFerm, 'tab:green')
axs[1, 3].set_title('Axis [1,0]')
axs[0, 4].plot(rLituus, 'tab:green')
axs[0, 4].set_title('Axis [1,0]')
axs[1, 4].plot(xLituus,yLituus, 'tab:green')
axs[1, 4].set_title('Axis [1,0]')


plt.setp(axs, xticks=[], yticks=[])
# axs[0,0].set_aspect('equal')
# plot curve and decorate with latex equation
# plt.plot(x, y)
# plt.plot(x, y)
plt.xlabel(r"$r(\theta)$")
plt.ylabel(r"$\theta^2$")
plt.title("Galilean spiral")
# plt.savefig("spiral.png")
plt.show()
```
