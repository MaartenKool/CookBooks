## Spirals

>Spiral, plane curve that, in general, winds around a point while moving ever farther from the point. Many kinds of spiral are known, the first dating from the days of ancient Greece. The curves are observed in nature, and human beings have used them in machines and in ornament, notably architectural‚Äîfor example, the whorl in an Ionic capital. (https://www.britannica.com/science/spiral-mathematics)

So in plain language a spiral is a circle with a varying radius. The way in which the radius is varied determined the type of spiral.

Fundamentally 2 types exist:

1. The radius changes at a constant linear pace. Often associated with Archemedes and Galileo. Every day examples include:
    - Roll of toilet paper, cello tape
    - Spring of a clock
    
    Typically human made
    

2. The radius changes logarithmically. Discovered by Descartes in 1638 and extended by Bernoulli:
    - Slug/snail shell
    - Thunderstorms seen from space
    
    Occurs in nature

### Achimedean Spirals

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
# use latex 
plt.rc('text', usetex=False)

#  define parameters
thetaMin, thetaMax = 0, 10*np.pi
steps = 500


# use numpy vectorization to generate plotting values
theta = np.linspace(thetaMin, thetaMax, steps)
#circle
r = np.pi
x = r*np.cos(theta)
y = r*np.sin(-theta)

#Archemedean Spiral
rArch= theta
xArch = rArch*np.cos(theta)
yArch = rArch*np.sin(-theta)

#Galilean Spiral
rGal= theta * theta * theta #last * theta for illustration purposes
xGal = rGal * np.cos(theta)
yGal = rGal * np.sin(-theta)

#Fermat Spiral
rFerm= np.sqrt(theta)
xFerm = rFerm*np.cos(theta)
yFerm = rFerm*np.sin(-theta)

#Lituus Spiral
rLituus = 1/rFerm
xLituus = rLituus*np.cos(theta)
yLituus = rLituus*np.sin(-theta)


fig, ax = plt.subplots(2, 5, figsize=(20,8))
ax[1, 0].plot([1,2,3,4,5],[r,r,r,r,r])
ax[1, 0].set_title('Radius')
ax[0, 0].plot(x, y,)
ax[0, 0].set_title('Plain Circle')
ax[1, 1].plot(rArch)
ax[1, 1].set_title('Radius')
ax[0, 1].plot(xArch, yArch)
ax[0, 1].set_title('Archemedean Spiral')
ax[1, 2].plot(rGal)
ax[1, 2].set_title('Radius')
ax[0, 2].plot(xGal,yGal)
ax[0, 2].set_title('Galilean Spiral')
ax[1, 3].plot(rFerm)
ax[1, 3].set_title('Radius')
ax[0, 3].plot(xFerm,yFerm)
ax[0, 3].set_title('Fermat Spiral')
ax[1, 4].plot(rLituus)
ax[1, 4].set_title('Radius')
ax[0, 4].plot(xLituus,yLituus)
ax[0, 4].set_title('Lituus Spiral')


plt.setp(ax, xticks=[], yticks=[])

fig.suptitle('A Circle and different types of spirals based on single dataset with different radi√Ø', fontsize=20)

# plt.savefig("spiral.png")
plt.show();
```

### Distance between 2 turns = 2 Pi * a

distance between turns 2 * pi * a

```python
3*pi*2
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
# use latex 
plt.rc('text', usetex=False)

# define parameters
thetaMin, thetaMax = 0, 36*np.pi
steps = 2000
a = 3
b = int(steps/4 -65  )

# use numpy vectorization to generate plotting values
theta = np.linspace(thetaMin, thetaMax, steps)
r = theta*a #*theta
x = r*np.cos(theta)
y = r*np.sin(-theta)

l=np.sqrt(np.square(x[b])+np.square(y[b]))
xnew= x[b]*1/l*(l+2*np.pi*a)
ynew= y[b]*1/l*(l+2*np.pi*a)

# fig, ax = plt.subplots(figsize=(8,8))
fig, ax = plt.subplots()


ax.set(ylim=[0,100], xlim=[0,100], autoscale_on=False)


ax.set_aspect('equal')
# plot curve and decorate with latex equation
# plt.plot(x[600:], y[600:])
# plt.scatter(x, y)
plt.plot(x, y, c='r')
plt.plot([x[b],xnew],[y[b],ynew])
plt.scatter(0,0)
plt.xlabel(r"$r(\theta)$")
plt.ylabel(r"$\theta$")
plt.title("Archemedes spiral")
# plt.savefig("spiral.png")
plt.show()
```
