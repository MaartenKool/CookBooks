## Spirals

>Spiral, plane curve that, in general, winds around a point while moving ever farther from the point. Many kinds of spiral are known, the first dating from the days of ancient Greece. The curves are observed in nature, and human beings have used them in machines and in ornament, notably architectural‚Äîfor example, the whorl in an Ionic capital. (https://www.britannica.com/science/spiral-mathematics)

So in plain language a spiral is a circle with a varying radius. The way in which the radius is varied determined the type of spiral.

Fundamentally 2 types exist:

1. The radius changes at a constant pace. Often associated with Archemedes and Galileo. Every day examples include:
    - Roll of toilet paper, cello tape
    - Spring of a clock
2. The radius changes logarithmically Discovered by Descartes in 1638 and extended by Bernoulli:
    - Slug/snail shell
    - Thunderstorms seen from space



### Achimedean Spirals

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
# use latex 
plt.rc('text', usetex=False)

#  define parameters
thetaMin, thetaMax = 0, 10*np.pi
steps = 500


# use numpy vectorization to generate plotting values
theta = np.linspace(thetaMin, thetaMax, steps)
#circle
r = np.pi
x = r*np.cos(theta)
y = r*np.sin(-theta)

#Archemedean Spiral
rArch= theta
xArch = rArch*np.cos(theta)
yArch = rArch*np.sin(-theta)

#Galilean Spiral
rGal= theta * theta * theta #last * theta for illustration purposes
xGal = rGal * np.cos(theta)
yGal = rGal * np.sin(-theta)

#Fermat Spiral
rFerm= np.sqrt(theta)
xFerm = rFerm*np.cos(theta)
yFerm = rFerm*np.sin(-theta)

#Lituus Spiral
rLituus = 1/rFerm
xLituus = rLituus*np.cos(theta)
yLituus = rLituus*np.sin(-theta)


fig, axs = plt.subplots(2, 5, figsize=(20,8))
axs[0, 0].plot([1,2,3,4,5],[r,r,r,r,r])
axs[0, 0].set_title('Radius')
axs[1, 0].plot(x, y,)
axs[1, 0].set_title('Plain Circle')
axs[0, 1].plot(rArch)
axs[0, 1].set_title('Radius')
axs[1, 1].plot(xArch, yArch)
axs[1, 1].set_title('Archemedean Spiral')
axs[0, 2].plot(rGal)
axs[0, 2].set_title('Radius')
axs[1, 2].plot(xGal,yGal)
axs[1, 2].set_title('Galilean Spiral')
axs[0, 3].plot(rFerm)
axs[0, 3].set_title('Radius')
axs[1, 3].plot(xFerm,yFerm)
axs[1, 3].set_title('Fermat Spiral')
axs[0, 4].plot(rLituus)
axs[0, 4].set_title('Radius')
axs[1, 4].plot(xLituus,yLituus)
axs[1, 4].set_title('Lituus Spiral')


plt.setp(axs, xticks=[], yticks=[])

# plt.xlabel(r"$r(\theta)$")
# plt.ylabel(r"$\theta^2$")
# plt.title("Galilean spiral")
# plt.savefig("spiral.png")
plt.show();
```
